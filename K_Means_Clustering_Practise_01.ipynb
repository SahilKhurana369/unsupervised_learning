```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
#configuration
D=2
K=3
N=300
```

```python
#Create data 
mul1 = np.array([0,0])
mul2 = np.array([5,5])
mul3 = np.array([0,5])
print(mul1)
print(mul2)
print(mul3)
```

```python
X = np.zeros((N, D))
```

```python
X
```

```python
X[:100, :] = np.random.randn(100, D) + mul1
```

```python
X[100:200, :] = np.random.randn(100, D) + mul2
```

```python
X[200:, :] = np.random.randn(100, D) + mul3
```

```python
X
```

```python
X.shape
```

```python
Y = np.array([0]*100 + [1]*100 + [2]*100)
```

```python
Y
```

```python
Y.shape
```

```python
#visulize the data
plt.scatter(X[:,0], X[:,1], c=Y);
```

```python
#how to get back a D-sized vector when taking the mean
X.mean(axis=0).shape
```

```python
#find the mean of each cluster
mean=np.zeros((K, D))
mean[0,:] = X[Y==0].mean(axis=0)
mean[1,:] = X[Y==1].mean(axis=0)
mean[2,:] = X[Y==2].mean(axis=0)
```

```python
#plot the mean with the data
plt.scatter(X[:,0],X[0:,1], c = Y)
plt.scatter(mean[:,0], mean[:,1], s=500, c='red', marker='*')
```
